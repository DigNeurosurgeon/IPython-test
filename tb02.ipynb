```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
```

```python
# import data
df = pd.read_csv('../data/tb.csv')

# select and prettify column names
dfs = df[['g_whoregion', 'country', 'year', 'new_sn', 'new_sp']]
dfs.columns = ['region', 'country', 'year', 'sn', 'sp']

# clean data
dfsc = dfs.dropna(subset = ['sp', 'sn'])
# dfsc.head()
# dfsc.ix[17]

# group by country
grouped = dfsc.groupby(['region', 'country', 'year']).agg(np.mean)
# grouped.head()

# work with African countries
# dfsc[dfsc.region == 'AFR'].groupby(['country']).agg(np.mean)

# select sn and plot
# first, plot one country throughout the years
sn1 = dfsc[dfsc.country == 'Afghanistan']
fig = plt.figure(figsize=(18,5))
# plt.plot(sn1.year, sn1.sn, label = 'Afghanistan')
# plt.show()
# plt.close()
# second, plot sn per country
countries = set(list(dfsc.country))
for c in countries:
    data = dfsc[dfsc.country == c]
    plt.plot(data.year, data.sn, label = c)
# plt.legend(loc=0)
plt.xlim(1995,2012)
plt.ylim(0,50000)
plt.show()

# sort by sn
# dfsc.sort(['sn'])
```
